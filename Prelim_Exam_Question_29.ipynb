Develop a python scripts that:

    Calculate total marks and percentage
    Assign grade for each subject based on marks:

90-100: A+

80-89: A

70-79: B+

60-69: B

50-59: C

40-49: D

Below 40: F (Fail)

    Determine overall result (Pass/Fail)



Performance Remarks:


Percentage ≥ 90: "Outstanding"

Percentage ≥ 80: "Excellent"

Percentage ≥ 70: "Very Good"

Percentage ≥ 60: "Good"

Percentage ≥ 40: "Satisfactory"

Below 40: "Needs Improvement"


```python
numofsubjects = int(input("Enter number of subjects: "))

total_marks = 0
max_marks = numofsubjects * 100

for x in range(1, numofsubjects + 1):
    marks = int(input(f"Enter marks for subject {x}: "))
    total_marks += marks

    if marks >= 90:
        grade = "A+"
    elif marks >= 80:
        grade = "A"
    elif marks >= 70:
        grade = "B+"
    elif marks >= 60:
        grade = "B"
    elif marks >= 50:
        grade = "C"
    elif marks >= 40:
        grade = "D"
    else:
        grade = "F"

    print(f"Grade for subject {x}: {grade}")

percentage = (total_marks / max_marks)*100

if percentage >= 40:
    result = "Pass"
else:
    result = "Fail"

if percentage >= 90:
    remark = "Outstanding"
elif percentage >= 80:
    remark = "Excellent"
elif percentage >= 70:
    remark = "Very Good"
elif percentage >= 60:
    remark = "Good"
elif percentage >= 40:
    remark = "Satisfactory"
else:
    remark = "Needs Improvement"

print("\nFinal Result")
print("Total Marks:", total_marks)
print("Percentage:", round(percentage, 2), "%")
print("Result:", result)
print("Performance Remark:", remark)
```
