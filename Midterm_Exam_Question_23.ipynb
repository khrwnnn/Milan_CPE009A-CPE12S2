Write a function that checks if a string of parentheses (), brackets [], and braces {} is balanced.

Examples:

    "({[]})" → True

    "([)]" → False

    "((()))" → True

    "({)}" → False


```python
def string_balanced(s):
    stack = []
    pairs = {')': '(', ']': '[', '}': '{'}

    for char in s:
        if char in '([{':
            stack.append(char)
        elif char in ')]}':
            if not stack or stack[-1] != pairs[char]:
                return False
            stack.pop()

    return len(stack) == 0



print(string_balanced("(({}))"))
print(string_balanced("([)]"))
print(string_balanced("((()))"))
print(string_balanced("{(})"))

```
