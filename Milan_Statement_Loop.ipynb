### Example 1 (For Loop Statement)

```python
week = ["Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]
for x in week:
  print(x)
```

### Example 2 (Break Statement in For Loop Statement)

```python
week = ["Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]
for x in week:
  print(x)
  if x == "Thursday":
    break
```

### Example 3 (Break Statement in For Loop Statement)

```python
week = ["Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]
for x in week:
  if x == "Thursday":
    break
print(x)
```

### Example 4 (Looping through a String)

```python
for x in "week":
  print(x)
```

### The range() Function

```python
for x in range(6):
  print(x)
for x in range(2, 6):
  print(x)
```

### Trial for range() Function

```python
for x in range(2, 6, 2):
  print(x)
```

### Nested Loops

```python
adjective = ["red","big","tasty"]
fruits = ["apple","banana","cherry"]
for x in adjective:
  for y in fruits:
    print(x, y)
```

### While Loop Statement

```python
i = 1
while i < 6:
  print(i)
  i += 1
  # i + 1 Assignment Operator
```

### The Break Statement in While Loop Statement

```python
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

### The Continue Statement in While Loop Statement

```python
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```

### Challenge for Continue Statement in While Loop Stat

```python
i = 0
while i < 6:
  i += 1
  if i == 3:
    print(i)
    continue
```

### The Else Statement in While Loop Statement

```python
i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("i is no longer less than 6")
```
