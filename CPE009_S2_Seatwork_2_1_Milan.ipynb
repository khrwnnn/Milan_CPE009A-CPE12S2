```python
# We want to create a python code for a grading scheme

# Create a function to accept a number of activities
# Compute for the score vs total
# Return the average grade of the student in %
def termCrit(actNumber):
    rawScore = []
    totalScore = []

    for i in range(actNumber):
        score = float(input(f"Score #{i+1}: "))
        total = float(input(f"Total #{i+1}: "))

        if score > total:
            print("Invalid score detected.")
            return 0

        rawScore.append(score)
        totalScore.append(total)

    grade = (sum(rawScore) / sum(totalScore)) * 100
    return grade
```

```python
# Create a function for the class standing per term
def termCS():
    print("\nHands-on Activities")
    hoa = termCrit(int(input("How many hands-on activities? ")))

    print("\nQuizzes")
    quiz = termCrit(int(input("How many quizzes? ")))

    print("\nAssignments")
    ass = termCrit(int(input("How many assignments? ")))

    grade = (0.5 * hoa) + (0.3 * quiz) + (0.2 * ass)
    return grade
```

```python
# Create a function to compute the term grade
def termGrade():
    print("\nComputing for the term grade.")
    examRaw = float(input("Exam Score: "))
    examTotal = float(input("Exam Total: "))
    examGrade = (examRaw / examTotal) * 100

    CSgrade = termCS()

    grade = (0.5 * examGrade) + (0.5 * CSgrade)
    return grade
```

```python
# Create a function to compute for the grades in the semester
def semesterGrade():
    print("\nPRELIM")
    prelim = termGrade()

    print("\nMIDTERM")
    midterm = termGrade()
    midFinal = (1/3 * prelim) + (2/3 * midterm)

    print("\nFINAL")
    final = termGrade()

    semesterFinal = (1/3 * midFinal) + (2/3 * final)
    return semesterFinal
```

```python
def convertGrade(finalGrade):
    if 99 <= finalGrade <= 100:
        return 1.00
    elif 96 <= finalGrade <= 98:
        return 1.25
    elif 93 <= finalGrade <= 95:
        return 1.50
    elif 90 <= finalGrade <= 92:
        return 1.75
    elif 87 <= finalGrade <= 89:
        return 2.00
    elif 84 <= finalGrade <= 86:
        return 2.25
    elif 81 <= finalGrade <= 83:
        return 2.50
    elif 78 <= finalGrade <= 80:
        return 2.75
    elif 75 <= finalGrade <= 77:
        return 3.00
    else:
        return 5.00
  #The class should add everything else
```
