```python
class Car:
    def __init__(self, name, mileage):
        self.name = name
        self.mileage = mileage

    def description(self):
        return f"The {self.name} car gives the mileage of {self.mileage} km/l"


car1 = Car("Toyota", 2000)
car2 = Car("Honda", 3000)

print(car1.description())
print(car2.description())
```

```python
class Animals:
    def __init__(self, name, habitat, temperature):
        self.name = name
        self.habitat = habitat
        self.temperature = temperature

    def feature(self):
        return f"The {self.name} which has a body temperature {self.temperature} lives in {self.habitat}"


mammals = Animals("Dog", "Land", "warm")
birds = Animals("Sparrow", "Air", "warm")
reptiles = Animals("Snake", "terrestrial", "cold")

#print(mammals.name)
#print(mammals.habitat)
print(mammals.feature())
```

```python

class Triangle:
  def __init__(self, base, height):
    self.name = "triangle"
    self.sides = 3
    self.base = base
    self.height = height

  def area(self):
    return (0.5 * self.base * self.height)

class Circle:
  def __init__(self, radius):
    self.name = "circle"
    self.sides = 0
    self.radius = radius

  def area(self):
    return (3.14 * self.radius * self.radius)

class Square:
  def __init__(self, side):
    self.name = "square"
    self.sides = 4
    self.side = side

  def area(self):
    return (self.side * self.side)

class Rectangle:
  def __init__(self, length, width):
    self.name = "rectangle"
    self.sides = 4
    self.lenght = length
    self.width = width

  def area(self):
    return (self.lenght * self.width)

t = Triangle(5, 4)
c = Circle(3)
s = Square(2)
r = Rectangle(6, 4)

print("Triangle area:", t.area())
print("Circle area:", c.area())
print("Square area:", s.area())
print("Rectangle area:", r.area())
```
