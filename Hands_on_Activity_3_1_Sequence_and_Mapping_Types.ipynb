List Methods and the Mutable Property of Lists

```python
numberlist2 = numberlist

print(numberlist)
print(numberlist2)

numberlist[0] = 6

print(numberlist)
print(numberlist2)

```

```python
numberlist2 = numberlist.copy()

print(numberlist2)

numberlist[0] = 5

print(numberlist)
print(numberlist2)
```

Dictionaries

```python
contactinfo = {
    'id':1,
    'first_name':'John',
    'last_name':'Doe',
    'contact_number':'09060611233'
}

print(contactinfo['id'])
print(contactinfo['first_name'])
print(contactinfo['last_name'])
print(contactinfo['contact_number'])

for k in contactinfo:
    print(k)

for k,v in contactinfo.items():
    print(k,v)

contactinfo['id'] = 2
contactinfo['first_name'] = 'Max'

print(contactinfo)
```

1. Distance Formula

```python
import math

coordinates_list = [(1,1),(2,3)]

x1,y1 = coordinates_list[0]
x2,y2 = coordinates_list[1]

distance = math.sqrt((x2-x1)**2 + (y2-y1)**2)

print("Distance:", distance)
```

3. Phonebook

```python
phonebook = [
    {'id':1,'name':'John','contact':'09123','address':'Manila'},
    {'id':2,'name':'Anna','contact':'09999','address':'Quezon City'}
]

search_id = int(input("Enter ID: "))

for person in phonebook:
    if person['id'] == search_id:
        print(person['name'])
        print(person['contact'])
        print(person['address'])
```
